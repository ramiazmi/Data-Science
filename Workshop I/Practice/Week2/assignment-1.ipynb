```python
def get_max_anagrams(ds):
    d = defaultdict(list)
    for word in ds:
        key = "".join(sorted(word))
        d[key].append(word)
        dl[key]+=1
    return d
```

```python
from collections import defaultdict
dl = {}
dl = defaultdict(lambda: 0, dl)
m = 0
wd = get_max_anagrams(wordclean)
m = max(dl.values())
# m = 3
r = defaultdict(list)
l = list()
for k,v in wd.items():
    if dl[k] == m:
        r[k] = v
        l.append((k+': '+ ', '.join(v)+' - '+str(m)))
%timeit
l
```

```python
def get_all_anagrams(ds):
    d = defaultdict(list)
    for word in ds:
        key = "".join(sorted(word))
        d[key].append(word)
    return d
```

```python
wd = get_all_anagrams(wordclean)
r = defaultdict(list)
l = list()
for k,v in wd.items():
    r[k] = v
    l.append((k+': '+ ', '.join(v)))
%timeit
l
```
