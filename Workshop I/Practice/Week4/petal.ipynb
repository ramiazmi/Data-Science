```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df_iris = pd.read_csv('iris-data.csv')
df_iris
```

```python
df = df_iris['petal_length_cm'] #[:50]
```

```python
def ecdf(ds):
    x = np.sort(ds)
    n = len(x)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
x_petal, y_petal = ecdf(df)
_= plt.plot(x_petal, y_petal, marker='.', linestyle='none')
plt.margins(0.02)
_= plt.xlabel('petal length cm')
_= plt.ylabel('ECDF')
plt.show()
```
