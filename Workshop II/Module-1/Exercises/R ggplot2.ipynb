**For the first few plots, use the mpg dataset**

```r
# install.packages('ggplot2', dep = TRUE)
# install.packages('ggthemes', dep = TRUE)
library(ggplot2)
library(ggthemes)
head(mpg)
```

**Histogram of hwy mpg values:**

```r
pl <- ggplot(mtcars)
pl + geom_histogram(binwidth=0.1,fill='red',alpha=0.4)
```

**Barplot of car counts per manufacturer with color fill defined by cyl count**

```r
g <- ggplot(mpg, aes(manufacturer))
g + geom_bar(aes(fill = cyl))
```
